```python
def permute(arr, n):
    result = []

    def backtrack(path):
        if len(path) == n:
            result.append(path[:])
            return
        for i in range(len(arr)):
                backtrack(path + [arr[i]])
    

    backtrack([])
    return result

input = 'ABCDEF'
n = 3
ans = permute(input, n) 

for i in ans:
    print(''.join(i))
```
