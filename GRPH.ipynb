```python
def fasta2dict(fil):
    dic = {}
    cur_scaf = ''
    cur_seq = []
    for line in open(fil):
        if line.startswith(">") and cur_scaf == '':
            cur_scaf = line.split(' ')[0]
        elif line.startswith(">") and cur_scaf != '':
            dic[cur_scaf[1:-1]] = ''.join(cur_seq)
            cur_scaf = line.split(' ')[0]
            cur_seq = []
        else:
            cur_seq.append(line.rstrip())
    dic[cur_scaf[1:-1]] = ''.join(cur_seq)
    return dic


dic = fasta2dict('/Users/yonghee_kim/Desktop/workspace/rosalind/rosalind_grph.txt')

ans = []
for id, seq in dic.items():
    for id2, seq2 in dic.items():
        if seq == seq2:
            continue
        if seq[-3:] == seq2[:3]:
            ans.append(id + ' ' + id2)

print("\n".join(ans))
```
