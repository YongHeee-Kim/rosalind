```python
def permute(arr):
    result = []
    used = [False] * len(arr)

    def backtrack(path):
        if len(path) == len(arr):
            result.append(path[:])
            return
        for i in range(len(arr)):
            if not used[i]:
                used[i] = True
                backtrack(path + [arr[i]])
                used[i] = False

    backtrack([])
    return result

n=6
ans = permute(list(range(1,n+1)))
print(len(ans))
for i in ans:
    print(*i)
```
