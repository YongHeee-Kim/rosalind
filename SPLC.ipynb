```python
def fasta2string(fil):
    lst = []
    cur_seq = []
    for line in open(fil):
        if line.startswith(">"):          
            lst.append(''.join(cur_seq))
            cur_seq = []
        else:
            cur_seq.append(line.rstrip())
    lst.append(''.join(cur_seq))
    return lst[1:]

rna_codon_table = {'UUU':'F','UUC':'F','UUA':'L','UUG':'L','CUU':'L','CUC':'L','CUA':'L',
                   'CUG':'L','AUU':'I','AUC':'I','AUA':'I','AUG':'M','GUU':'V','GUC':'V',
                   'GUA':'V','GUG':'V','UCU':'S','UCC':'S','UCA':'S','UCG':'S','AGU':'S',
                   'AGC':'S','CCU':'P','CCC':'P','CCA':'P','CCG':'P','ACU':'T','ACC':'T',
                   'ACA':'T','ACG':'T','GCU':'A','GCC':'A','GCA':'A','GCG':'A','UAU':'Y',
                   'UAC':'Y','CAU':'H','CAC':'H','CAA':'Q','CAG':'Q','AAU':'N','AAC':'N',
                   'AAA':'K','AAG':'K','GAU':'D','GAC':'D','GAA':'E','GAG':'E','UGU':'C',
                   'UGC':'C','UGG':'W','CGU':'R','CGC':'R','CGA':'R','CGG':'R','AGA':'R',
                   'AGG':'R','GGU':'G','GGC':'G','GGA':'G','GGG':'G','UAA':'Stop','UAG':'Stop','UGA':'Stop'}


seq = fasta2string('')
ans = ''

seq[0] = seq[0].replace('T', 'U')

for i in seq[1:]:
    i = i.replace('T', 'U')
    seq[0] = seq[0].replace(i,'')


for i in range(0,len(seq[0])-3, 3):
    ans +=rna_codon_table[seq[0][i:i+3]]


print(ans)
```
