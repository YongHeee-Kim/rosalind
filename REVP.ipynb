```python
def fasta2string(fil):
    lst = []
    cur_seq = []
    for line in open(fil):
        if line.startswith(">"):          
            lst.append(''.join(cur_seq))
            cur_seq = []
        else:
            cur_seq.append(line.rstrip())
    lst.append(''.join(cur_seq))
    return lst[1:][0]

def check(seq):
    halflen = len(seq)//2
    dic = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
    a = ''.join(dic[a] for a in seq[:halflen])
    b = seq[halflen:]
    if a==b[::-1]:
        return 1
    return 0

seq = fasta2string('')
ans=[]
for i in range(len(seq)-3):
    for ii in range(4, 13, 2):
        if check(seq[i:i+ii]) == 1 and i+ii <=len(seq):
            ans.append(str(i+1)+" " + str(ii))
           

for i in ans:
    print(i)
```
