```python
def fasta2string(fil):
    lst = []
    cur_seq = []
    for line in open(fil):
        if line.startswith(">"):          
            lst.append(''.join(cur_seq))
            cur_seq = []
        else:
            cur_seq.append(line.rstrip())
    lst.append(''.join(cur_seq))
    return lst[1:][0]


def DNA2RNA(dnaseq):
    dic = {'A':'U', 'U':'A', 'C':'G', 'G':'C'}
    rnaseq = ''.join(dic[a] for a in dnaseq)

    return rnaseq

rna_codon_table = {'UUU':'F','UUC':'F','UUA':'L','UUG':'L','CUU':'L','CUC':'L','CUA':'L',
                   'CUG':'L','AUU':'I','AUC':'I','AUA':'I','AUG':'M','GUU':'V','GUC':'V',
                   'GUA':'V','GUG':'V','UCU':'S','UCC':'S','UCA':'S','UCG':'S','AGU':'S',
                   'AGC':'S','CCU':'P','CCC':'P','CCA':'P','CCG':'P','ACU':'T','ACC':'T',
                   'ACA':'T','ACG':'T','GCU':'A','GCC':'A','GCA':'A','GCG':'A','UAU':'Y',
                   'UAC':'Y','CAU':'H','CAC':'H','CAA':'Q','CAG':'Q','AAU':'N','AAC':'N',
                   'AAA':'K','AAG':'K','GAU':'D','GAC':'D','GAA':'E','GAG':'E','UGU':'C',
                   'UGC':'C','UGG':'W','CGU':'R','CGC':'R','CGA':'R','CGG':'R','AGA':'R',
                   'AGG':'R','GGU':'G','GGC':'G','GGA':'G','GGG':'G','UAA':'Stop','UAG':'Stop','UGA':'Stop'}

dnaseq= fasta2string('/Users/yonghee_kim/Desktop/workspace/rosalind/rosalind_orf.txt')

seq = dnaseq.replace('T', 'U')
r_seq = seq[::-1]
r_seq = DNA2RNA(r_seq)
start_codon = "AUG"
ans = []

indices = [i for i in range(len(seq)-2) if seq[i:i+3] == start_codon]
indicess = [i for i in range(len(seq)-2) if seq[i:i+3] == 'UAA' or seq[i:i+3] == 'UAG' or seq[i:i+3] == 'UGA' ]

r_indices = [i for i in range(len(r_seq)-2) if r_seq[i:i+3] == start_codon]
r_indicess = [i for i in range(len(r_seq)-2) if r_seq[i:i+3] == 'UAA' or r_seq[i:i+3] == 'UAG' or r_seq[i:i+3] == 'UGA' ]

for i in indices:
    for ii in indicess:
        if (ii-i) % 3 == 0 and ii-i> 0:
            ans_s = seq[i:ii]
            ans.append(''.join(rna_codon_table[ans_s[i:i+3]] for i in range(0, len(ans_s)-2, 3)))
            break

for i in r_indices:
    for ii in r_indicess:
        if (ii-i) % 3 == 0 and ii-i> 0:
            ans_s = r_seq[i:ii]
            ans.append(''.join(rna_codon_table[ans_s[i:i+3]] for i in range(0, len(ans_s)-2, 3)))
            break


for i in list(set(ans)):
    print(i)
```
