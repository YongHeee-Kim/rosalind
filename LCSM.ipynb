```python
def fasta2lst(fil):
    lst = []
    cur_seq = []
    for line in open(fil):
        if line.startswith(">"):          
            lst.append(''.join(cur_seq))
            cur_seq = []
        else:
            cur_seq.append(line.rstrip())
    lst.append(''.join(cur_seq))
    return lst[1:]

def common_substrings(strings, min_len=2):
    base = strings[0]
    ans = ''

    for i in range(len(base)):
        for j in range(i + min_len, len(base) + 1):
            sub = base[i:j]            
            if all(sub in s for s in strings[1:]):
                if len(sub) > len(ans):
                    ans = sub
    return ans

file_path = ''
seq = fasta2lst(file_path)

print(common_substrings(seq, min_len=2))
```
