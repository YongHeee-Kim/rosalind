```python
import requests
from Bio import SeqIO
from io import StringIO

with open("/Users/yonghee_kim/Desktop/workspace/rosalind/rosalind_mprt-2.txt", "r") as f:
    text = f.read()
proteins=text.split("\n")[:-1]
print(proteins)
url = 'http://www.uniprot.org/uniprot/'
ans = []

for name in proteins:
    response = requests.get(url+name.split("_")[0]+'.fasta')

    fasta_io = StringIO(response.text)
    record = SeqIO.read(fasta_io, "fasta")
    seq = record.seq
    
    for i in range(len(seq)-3):
        if seq[i] == 'N' and seq[i+1] != 'P' and seq[i+2] in ("S", "T") and seq[i+3] != 'P':
            ans.append(i+1)
    if ans != []:
        print(name)
        print(*ans)
    ans = []
```
