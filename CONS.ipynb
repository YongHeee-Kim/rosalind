```python
path = " "

with open(path, "r", encoding="utf-8") as f:
    text = f.read()
num = text.count('_')

text = "".join(
    line.strip()
    for line in text.splitlines()
    if not line.startswith(">")
)

c_len = len(text)//num
library = {'A':[0]*(c_len), 'C':[0]*(c_len), 'G':[0]*(c_len), 'T':[0]*(c_len)}
ans = ''
anss = ''

for i in range(c_len):
    for ii in range(num):
        library[text[ii*c_len+i]][i] += 1
    ans = ans + max(library, key = lambda k: library[k][i])

anss +="\n".join(
    f"{k}: " + " ".join(map(str, v))
    for k, v in library.items()
)

print(ans + '\n' +anss)

```
